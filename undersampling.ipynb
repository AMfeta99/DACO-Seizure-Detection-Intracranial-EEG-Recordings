Due to the high amount of signals on the dataset, the computational cost wasn't viable for working on a short amount of time and with limited resources.
This notebook takes the dataset after preprocessing and randomly undersamples it, keeping the ratio preictal:interictal equal to 1:2, without eliminating preictal signals.

```
def appendLabel(labels, filename): 
  if 'interictal' in filename:
    labels.append(0)
  else:
    labels.append(1)
  return labels
```

```
import os
import scipy.io

X_dog1 = [];
y_dog1 = [];
X_dog2 = [];
y_dog2 = [];
X_dog3 = [];
y_dog3 = [];
X_dog5 = [];
y_dog5 = [];
X_patient1 = [];
y_patient1 = [];
X_patient2 = [];
y_patient2 = [];

for dirname, _, filenames in os.walk('/content/drive/MyDrive/stft12020/train'):
   for filename in filenames:

     if 'Dog_1' in filename:
       X_dog1.append(filename)
       y_dog1 = appendLabel(y_dog1, filename)

     elif 'Dog_2' in filename:
       X_dog2.append(filename)
       y_dog2 = appendLabel(y_dog2, filename)   

     elif 'Dog_3' in filename:
       X_dog3.append(filename)
       y_dog3 = appendLabel(y_dog3, filename)

     elif 'Dog_5' in filename:
       X_dog5.append(filename)
       y_dog5 = appendLabel(y_dog5, filename)

     elif 'Patient_1' in filename:
       X_patient1.append(filename)
       y_patient1 = appendLabel(y_patient1, filename)

     elif 'Patient_2' in filename:
       X_patient2.append(filename)
       y_patient2 = appendLabel(y_patient2, filename)
```

```
def undersampler(X, y):

  num_preictal = min(list(Counter(y).values()))

  X = np.asarray(X).reshape(-1, 1)
  y = np.asarray(y)

  rus = RandomUnderSampler(random_state=0, ratio={0: 2*num_preictal, 1: num_preictal})
  X_resampled, y_resampled = rus.fit_sample(X, y)

  return X_resampled, y_resampled
```

```
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter
import numpy as np

X_dog1_resampled, y_dog1_resampled = undersampler(X_dog1, y_dog1)
X_dog2_resampled, y_dog2_resampled = undersampler(X_dog2, y_dog2)
X_dog3_resampled, y_dog3_resampled = undersampler(X_dog3, y_dog3)
X_dog5_resampled, y_dog5_resampled = undersampler(X_dog5, y_dog5)
X_patient1_resampled, y_patient1_resampled = undersampler(X_patient1, y_patient1)
X_patient2_resampled, y_patient2_resampled = undersampler(X_patient2, y_patient2)
```
